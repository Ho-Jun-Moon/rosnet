### rosnet.neural_network.math

```python
def mat_outer(mat1, mat2, axes = (0, 0)):
  def get_vector(mat, axis, n):
    if axis == 0:
      return mat[n]
    else:
      return mat.T[n]

  mat_arr = []
  I = mat1.shape[axes[0]]

  for i in range(I):
    v1 = get_vector(mat1, axes[0], i)
    v2 = get_vector(mat2, axes[1], i)
    mat_arr.append(np.outer(v1, v2))

  return np.array(mat_arr)
```
